```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import matplotlib
```

```python
df = pd.read_excel('usuarios.xlsx',sheet_name='Datos')
df = df.iloc[:,[1,2]].dropna()
df.columns = ['Año','Usuarios']
df['Año'] = [i[0] for i in df['Año'].str.split('*')]
```

```python
def add_value_label(x_list,y_list):
    for i in range(0, len(x_list)):
        plt.text(i-0.18,y_list[i]+0.08,y_list[i],fontsize=14,)
        # print(i)
# df
```

```python
current_palette = matplotlib.colors.hex2color('#a100ff')

plt.suptitle('Número de usuarios de realidad aumentada móvil a\nnivel mundial desde 2019 hasta 2024',ha='left',x=0,y=1.02,fontsize=14, weight='bold')  # y=-0.02  Titulo abajo
sns.barplot(data=df,x='Año',y='Usuarios',color=current_palette)
add_value_label(df['Año'].to_list(),df['Usuarios'].to_list())
for i in ['bottom','left','top','right']:
  plt.gca().spines[i].set_linewidth(0)
plt.xlabel('Año',fontsize=14)
plt.ylabel('Usuarios - Miles de Millones',fontsize=14)
plt.savefig('Grafico1.png',dpi=300, bbox_inches='tight')
plt.show()
```

```python
df = pd.read_excel('industria.xlsx',sheet_name='Datos')
df = df.iloc[:,[1,2,3]].dropna()
df.columns = ['Industria','2020','2025*']
df['2025*'] = df['2025*'].astype('float')
```

```python
melt = pd.melt(df, id_vars='Industria',var_name='Año',value_name='Usuarios')
```

```python
palette = {'2025*': '#a100ff',
           '2020':'#d7a5d2'
}
# plt.figure(figsize=(7,5))
plt.suptitle('Número de Usuarios de RA y RV 2020 - 2025*\na Nivel Mundial por Industria',y=1.02,x=0.13,ha='left',fontsize=14,weight='bold')
sns.barplot(data=melt,y='Industria',x='Usuarios',hue='Año',palette=palette)
labels = list(df.sort_values(by='2025*',ascending=False)['Industria'].values)
for i in ['bottom','left','top','right']:
  plt.gca().spines[i].set_linewidth(0)
plt.yticks(ticks=[0,1,2,3,4,5,6,7],labels=labels,fontsize=12)
plt.ylabel('Industria',fontsize=14)
plt.xlabel('Usuarios RA y RV - Millones',fontsize=14,)
plt.legend(loc='lower right')
plt.savefig('Grafico2.png',dpi=300, bbox_inches='tight')
plt.show()
# plt.x
```

```python
df = pd.read_csv('remodelaciones.csv')
df.reset_index(inplace=True)
df = df.drop(0)
df.columns = ['Fecha','Interes']
df.head(2)
df['Fecha'] = pd.to_datetime(df['Fecha'])
df['Interes'] = df['Interes'].astype('int')
```

```python
plt.suptitle('Ineteres en el Termino "Remodelación"\nSegun Google Trends 2018 - 2023',y=1.02,x=0.13,ha='left',fontsize=14,weight='bold' )
sns.lineplot(data=df,x='Fecha',y='Interes',color='#a100ff')
for i in ['bottom','left','top','right']:
  plt.gca().spines[i].set_linewidth(0)
plt.grid(axis='y')
plt.ylabel('Indice Interes\n Google Trends',fontsize=14)
plt.xlabel('Fecha',fontsize=14)
plt.savefig('Grafico3.png',dpi=300, bbox_inches='tight')
plt.show()
```
